```python
import matplotlib.pyplot as plt
import numpy as np
import copy
```

```python
with open("../Data/the_icy_lake.txt") as file:
    raw_lake = file.read()
```

```python
lake = raw_lake.split('\n')
icy_lake = np.asarray([np.asarray([int(x) for x in row.split('\t')]) for row in lake])
plt.imshow(icy_lake)
```

```python
#Reading path from java code

with open("../Data/two_random_paths") as file:
    paths = file.read()

paths = paths.split('\n\n')
for i in range(len(paths)):
    paths[i] = [[int(xy.split(' ')[0]), int(xy.split(' ')[1])] for xy in paths[i].split('\n')]

path_img = copy.deepcopy(icy_lake)
for i in range(len(paths)):
    for x, y in paths[i]:
        path_img[y][x] = 2+2*i

plt.imshow(path_img)
```
